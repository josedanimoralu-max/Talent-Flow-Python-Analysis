## Propuesta 2: Relación entre satisfacción laboral y antigüedad
Variables involucradas: JobSatisfaction, EnvironmentSatisfaction, YearsAtCompany, YearsInCurrentRole.  
Pregunta guiadora: *¿Hay relación entre el tiempo que lleva un empleado en su puesto y su nivel de satisfacción laboral?*

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

print("Conteo Attrition:")
print(df['Attrition'].value_counts())
print("\nConteo Attrition_Flag:")
print(df['Attrition_Flag'].value_counts())
```

```python
# ============================================================
# CELDA 2 — ANÁLISIS PROFUNDO: ANTIGÜEDAD
# ============================================================

# Estadísticos descriptivos por Attrition
antiguedad_stats = df.groupby('Attrition')['YearsAtCompany'].describe().round(2)
print("[Antigüedad] Estadísticos por Attrition:")
print(antiguedad_stats)

# Tramos de antigüedad (0–2, 3–5, 6–10, 11+)
bins_ant = [0, 2, 5, 10, 40]
labels_ant = ['0-2 años', '3-5 años', '6-10 años', '11+ años']

df['YearsAtCompany_bin'] = pd.cut(
    df['YearsAtCompany'],
    bins=bins_ant,
    labels=labels_ant,
    include_lowest=True
)

# Tasa de renuncia por tramo
tasa_attr_por_ant = (
    df.groupby('YearsAtCompany_bin')['Attrition_Flag']
      .mean()
      .reset_index()
      .rename(columns={'Attrition_Flag': 'AttritionRate'})
)

print("\n[Tasa de renuncia por tramo de antigüedad]")
print(tasa_attr_por_ant)

# Gráfico
plt.figure(figsize=(8, 5))
sns.lineplot(
    data=tasa_attr_por_ant,
    x='YearsAtCompany_bin',
    y='AttritionRate',
    marker='o'
)
plt.title('Tasa de renuncia según antigüedad en la empresa')
plt.xlabel('Tramo de antigüedad')
plt.ylabel('Tasa de Attrition')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ============================================================
# CELDA 3 — ANÁLISIS PROFUNDO: HORAS EXTRA (OVERTIME)
# ============================================================

# Tasa de renuncia por OverTime
tasa_attr_por_ot = (
    df.groupby('OverTime')['Attrition_Flag']
      .mean()
      .reset_index()
      .rename(columns={'Attrition_Flag': 'AttritionRate'})
)

print("[Tasa de renuncia por OverTime]")
print(tasa_attr_por_ot)

# Gráfico
plt.figure(figsize=(6, 5))
sns.barplot(
    data=tasa_attr_por_ot,
    x='OverTime',
    y='AttritionRate'
)
plt.title('Probabilidad de renuncia según OverTime')
plt.xlabel('Realiza horas extra')
plt.ylabel('Tasa de Attrition')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# ============================================================
# CELDA 4 — ANÁLISIS PROFUNDO: INGRESO MENSUAL
# ============================================================

# Estadísticos por Attrition
income_stats = df.groupby('Attrition')['MonthlyIncome'].describe().round(2)
print("[Ingresos] Estadísticos por Attrition:")
print(income_stats)

# Tramos salariales por cuartiles
df['Income_bin'] = pd.qcut(
    df['MonthlyIncome'],
    q=4,
    labels=['Muy bajo', 'Bajo', 'Medio', 'Alto']
)

# Tasa de renuncia por nivel salarial
tasa_attr_por_ing = (
    df.groupby('Income_bin')['Attrition_Flag']
      .mean()
      .reset_index()
      .rename(columns={'Attrition_Flag': 'AttritionRate'})
)

print("\n[Tasa de renuncia por tramo de ingreso]")
print(tasa_attr_por_ing)

# Gráfico
plt.figure(figsize=(8, 5))
sns.barplot(
    data=tasa_attr_por_ing,
    x='Income_bin',
    y='AttritionRate'
)
plt.title('Tasa de renuncia según nivel de ingreso')
plt.xlabel('Tramo de ingreso')
plt.ylabel('Tasa de Attrition')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# ============================================================
# CELDA 5 — INTERACCIÓN: ANTIGÜEDAD × OVERTIME
# ============================================================

tabla_ant_ot = pd.pivot_table(
    df,
    values='Attrition_Flag',
    index='YearsAtCompany_bin',
    columns='OverTime',
    aggfunc='mean'
)

print("[Attrition por Antigüedad × OverTime]")
print(tabla_ant_ot.round(3))

plt.figure(figsize=(7, 5))
sns.heatmap(
    tabla_ant_ot,
    annot=True,
    fmt=".3f",
    cmap='Blues'
)
plt.title('Tasa de Attrition por Antigüedad × OverTime')
plt.xlabel('OverTime')
plt.ylabel('Tramo de antigüedad')
plt.show()
```

```python
# ============================================================
# CELDA 6 — INTERACCIÓN: OVERTIME × INGRESO
# ============================================================

tabla_ot_ing = pd.pivot_table(
    df,
    values='Attrition_Flag',
    index='Income_bin',
    columns='OverTime',
    aggfunc='mean'
)

print("[Attrition por Ingreso × OverTime]")
print(tabla_ot_ing.round(3))

plt.figure(figsize=(7, 5))
sns.heatmap(
    tabla_ot_ing,
    annot=True,
    fmt=".3f",
    cmap='Blues'
)
plt.title('Tasa de Attrition por Ingreso × OverTime')
plt.xlabel('OverTime')
plt.ylabel('Tramo de ingreso')
plt.show()
```

```python
# ============================================================
# CELDA 8 — INTERPRETACIÓN AUTOMÁTICA (OPCIONAL)
# ============================================================

print("\n===== INTERPRETACIÓN PROFESIONAL =====\n")

print("1) ANTIGÜEDAD:")
print("   - La tasa de Attrition es mayor en los primeros 2 años.")
print("   - Disminuye notablemente después del año 3.")
print("   - Esto indica un riesgo temprano en el ciclo laboral.\n")

print("2) HORAS EXTRA (OverTime):")
print("   - La renuncia es considerablemente más alta entre quienes realizan OverTime.")
print("   - Es un factor de riesgo directo y consistente.\n")

print("3) INGRESO MENSUAL:")
print("   - Los tramos salariales bajos concentran mayor Attrition.")
print("   - La probabilidad disminuye a medida que aumenta el ingreso.\n")

print("4) INTERACCIONES:")
print("   - Antigüedad × OverTime: los empleados nuevos que hacen OverTime son el grupo de mayor riesgo.")
print("   - Ingreso × OverTime: el salario modera el impacto del OverTime, pero no lo elimina.\n")


print("En conjunto, antigüedad baja, carga laboral alta y salario bajo forman un perfil multifactorial de riesgo de renuncia.")
```

```python
# ============================================================
# CELDA 5 — Distribución salarial (KDE) por Attrition
# ============================================================

plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='MonthlyIncome', hue='Attrition', fill=True, common_norm=False, alpha=0.4)
plt.title('Distribución salarial de empleados que renuncian vs permanecen')
plt.xlabel('Ingreso mensual')
plt.ylabel('Densidad')
plt.show()
```

```python
# ============================================================
# CELDA 6 — Brecha salarial (%)
# ============================================================

ingreso_no = df[df['Attrition_Flag']==0]['MonthlyIncome'].mean()
ingreso_yes = df[df['Attrition_Flag']==1]['MonthlyIncome'].mean()

brecha_abs = ingreso_no - ingreso_yes
brecha_pct = (brecha_abs / ingreso_no) * 100

print(f"Ingreso promedio - Permanecen: {ingreso_no:.2f}")
print(f"Ingreso promedio - Renuncian: {ingreso_yes:.2f}")
print(f"Brecha absoluta: {brecha_abs:.2f}")
print(f"Brecha porcentual: {brecha_pct:.2f}% menos ganan quienes renuncian")
```

```python
# ============================================================
# CELDA 7 — Interacción Ingreso × Antigüedad
# ============================================================

tabla_ing_ant = pd.pivot_table(
    df,
    values='Attrition_Flag',
    index='Income_bin',
    columns='YearsAtCompany_bin',
    aggfunc='mean'
)

print("=== Attrition por Ingreso × Antigüedad ===")
print(tabla_ing_ant.round(3))

plt.figure(figsize=(9, 6))
sns.heatmap(tabla_ing_ant, annot=True, fmt=".3f", cmap="Blues")
plt.title('Tasa de Attrition — Interacción Ingreso × Antigüedad')
plt.xlabel('Tramo de antigüedad')
plt.ylabel('Tramo salarial')
plt.show()
```
